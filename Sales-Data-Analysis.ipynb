```python
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])
```

```python
df["Month"] = df["Sale_Date"].dt.month
df["Year"] = df["Sale_Date"].dt.year
```

```python
df['Sale_Date']
```

```python
#Total Sales
total_sales = df["Sales_Amount"].sum()
print("Total Sales:", total_sales)
```

```python
#Total sales based on region
region_sales = df.groupby("Region")["Sales_Amount"].sum()
print(region_sales)
```

```python
#Sales by Product Category
category_sales = df.groupby("Product_Category")["Sales_Amount"].sum()
print(category_sales)
```

```python
#Top 5 Sales Representatives
top_rep = df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=False).head()
print(top_rep)
```

```python
#Sales by Customer Type
customer_sales = df.groupby("Customer_Type")["Sales_Amount"].sum()
print(customer_sales)
```

```python
#Profit Calculation
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

profit_by_category = df.groupby("Product_Category")["Profit"].sum()
print(profit_by_category)
```

```python
#Visualization
category_sales.plot(kind='bar')
plt.title("Sales by Product Category")
plt.show()
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```

```python
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()

monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
#Highest sales region
df.groupby("Region")["Sales_Amount"].sum().sort_values(ascending=False)
```

```python
#Most profitable Category
df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]

df.groupby("Product_Category")["Profit"].sum().sort_values(ascending=False)
```

```python
#Top sales representative
df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=False)
```

```python
#Peak sales month
df["Month"] = df["Sale_Date"].dt.month
df.groupby("Month")["Sales_Amount"].sum().sort_values(ascending=False)
```
